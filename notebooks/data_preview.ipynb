## Как будут выглядеть колонки после нормализации

```python
import re

def show_normalized_cols(df):
    for col in df.columns:
        new = re.sub(r"[^a-z0-9]+", "_", col.lower()).strip("_")
        print(f"  {col!r:45s} -> {new!r}")

print("accidents:")
show_normalized_cols(accidents)
print()
print("events:")
show_normalized_cols(events)
```
