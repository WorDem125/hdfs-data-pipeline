```python
import pandas as pd

road = pd.read_csv("../data/raw/Road_Accident_Data.csv")
events = pd.read_csv("../data/raw/World_Important_Dates.csv")

from IPython.display import display

print("Road Accident Data")
display(road.head())

print("World Important Dates")
display(events.head())
```

```python
print("=== ROAD INFO ===")
print(road.info())

print("\n=== EVENTS INFO ===")
print(events.info())

print("\n=== NULL VALUES ROAD ===")
print(road.isnull().sum())

print("\n=== NULL VALUES EVENTS ===")
print(events.isnull().sum())

print("\n=== DUPLICATES ===")
print("Road:", road.duplicated().sum())
print("Events:", events.duplicated().sum())
```
